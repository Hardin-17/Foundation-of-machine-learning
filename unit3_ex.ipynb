```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Spotify_Hours":[1,2,3,4,6,2,3,4,6,3,2,3,4,6,2,3,4,6,3,2],
    "YT_Music":[1,1,2,2,3,1,2,3,3,2,1,2,3,3,1,2,3,3,2,1],
    "Podcast_Hours":[0,1,1,2,2,1,1,2,2,1,0,1,2,2,1,1,2,2,1,0],
    "Work_Hours":[6,7,8,7,6,7,8,7,6,7,8,7,6,7,8,7,6,7,8,7],
    "Productivity":[6,7,8,7,6,7,8,7,6,7,8,7,6,7,8,7,6,7,8,7]
}

df = pd.DataFrame(data)

# Scatter Plot
plt.scatter(df["Spotify_Hours"], df["Productivity"])
plt.xlabel("Spotify Hours")
plt.ylabel("Productivity")
plt.title("Music vs Productivity")
plt.show()

# Histogram
plt.hist(df["Podcast_Hours"], bins=5)
plt.xlabel("Podcast Hours")
plt.ylabel("Frequency")
plt.title("Podcast Listening Distribution")
plt.show()

# Bar Chart
avg_productivity = df.groupby("Work_Hours")["Productivity"].mean()
avg_productivity.plot(kind="bar")
plt.xlabel("Work Hours")
plt.ylabel("Average Productivity")
plt.title("Work Hours vs Productivity")
plt.show()
```

INFERENCE: 
The analysis suggests that balanced digital media consumption, especially music, may positively support productivity, while moderate podcast listening is common among users. Additionally, maintaining regular work hours is associated with higher productivity levels.
