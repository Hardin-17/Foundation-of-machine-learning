DATA VISUALISATION

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data={
    "Name":["Aman","riya","Rahul","Sneha","Arjun","Priya","Karan","Neha"],
    "Age":[18,17,20,15,18,19,19,15],
    "Gender":["Male","Female","Male","Female","Male","Female","Male","Female"],
    "Marks":[89,64,78,98,97,65,100,65]
}
df = pd.DataFrame(data)
```

```python
#uni-variate analysis
plt.plot(df["Marks"])
plt.title("Marks Trends")
plt.xlabel("Student Name")
plt.ylabel("Marks")
```

```python
#bi-variate 
plt.bar(df["Name"],df["Marks"])
plt.title("Marks of students")
plt.xlabel("Students name")
plt.ylabel("Marks")
plt.show()
```

```python
plt.hist(df["Age"],bins=5)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

Scatterplot
Relationship with age and marks

```python
plt.scatter(df["Age"],df["Marks"])
plt.title("Age vs marks")
plt.xlabel("Age")
plt.ylabel("Marks")
plt.show()
```

pie-chart
gender distribution

```python
gender_counts=df["Gender"].value_counts()
plt.pie(gender_counts,labels=gender_counts.index,autopct="%1.1f%%")
plt.title("Gender Distribution")
plt.show()
```

Customizing plot
example with colours and grid

```python
plt.bar(df["Name"],df["Marks"],color="black")
plt.title("Students vs marks")
plt.xlabel("Student name")
plt.ylabel("Marks")
plt.grid(True)
plt.show()
```

Subplot
show two graphs together

```python
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.hist(df["Age"])
plt.title("Age Distribution")

plt.subplot(1,2,2)
plt.bar(df["Name"],df["Marks"])
plt.title("marks of students")
plt.show()
```
