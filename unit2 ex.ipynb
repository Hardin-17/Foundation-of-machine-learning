```python
import pandas as pd

df = pd.DataFrame({
    "Name": ["Aman", "Saurav", None, "Shubham", "Abhishek",
             None, "Rohit", "Priya", "Neha", "Arjun"],
    
    "Department": ["IT", "HR", "Finance", "IT", "Marketing",
                   "HR", "Finance", "IT", "Marketing", "HR"],
    
    "Role": ["Developer", "Recruiter", "Accountant", "Tester", "SEO Specialist",
             "HR Manager", "Financial Analyst", "Accountant", "Tester", "HR Executive"],
    
    "Experience": [2, 5, 3, 0, 4, 6, 1, 0, 2, 5],
    
    "Salary": [40000, 55000, 50000, 35000, 45000,
               60000, 48000, 42000, 39000, 52000],
    
    "Score": [85, 78, 88, None,90, 76, 84, None, 75, 89]
})

df
```

```python
df.groupby("Department")["Salary"].mean()
```

```python
df.groupby("Department")["Salary"].mean()
```

```python
df.groupby("Department")["Score"].count()
```

```python
df.groupby("Department").size()
```

```python
df.groupby('Department')['Salary'].agg(['mean','max','min'])
```

```python
df.groupby('Department')['Score'].agg(['mean','max','min','count'])
```

```python
df.groupby(["Department","Role"])["Salary"].mean()
```
