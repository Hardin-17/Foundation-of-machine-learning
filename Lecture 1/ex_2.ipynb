task5

```python
row_avg = np.mean(arr, axis=1)

grades = np.where(
    row_avg > 80, "Grade A",
    np.where(row_avg > 70, "Grade B", "No Grade")
)

print("Row averages:", row_avg)
print("Grades:", grades)
```
