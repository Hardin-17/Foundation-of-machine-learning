where in numpy

```python
arr=np.array([10,-5,20,-3,15])
```

```python
res=np.where(arr<0,0,arr)
print(res)
```

```python
arr=np.array([1,2,3,4,5])
```

```python
res=np.where(arr%2!=0,-1,arr)
print(res)
```

```python
marks=np.array([1,2,3,4,50])
```

```python
stat=np.where(marks>=50,1,0)
print(stat)
```

```python
data=np.array([10,20,30,40])
```

```python
mean_val=np.mean(data)
result=np.where(data>mean_val,mean_val,data)
print(result)
```
